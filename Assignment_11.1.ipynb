```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
print("Reading from the given location")
print(df)
df.info()

males = (df['sex'].str[0].str.lower() == 'm').sum()
females = (df['sex'].str[0].str.lower() == 'f').sum()
print("Printing Males")
print(males)
print("Printing Females")
print(females)

# Data to plot
labels = 'Male', 'Female'
malesFemales = [males, females]
colors = ['yellowgreen', 'lightcoral']
explode = (0.1, 0)  # explode 1st slice
 
# Plot
plt.pie(malesFemales, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()

faresByAge = df.dropna(subset=['age'])
faresByAge = faresByAge.loc[:, faresByAge.columns.isin(['age','fare', 'sex'])]
print("Printing faresByAge")
print(faresByAge)

mapping = {'male' : 'blue', 'female' : 'red'}
plt.scatter(faresByAge['age'], faresByAge['fare'], alpha=0.5, c=faresByAge['sex'].map(mapping))
plt.title('Scatter plot of fares  by Age (Male and Females)')
plt.xlabel('Age---->')
plt.ylabel('Fare----->')
plt.show()
```
